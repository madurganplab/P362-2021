# I-V curves of active elements

In this section we characterize a simple circuit, the voltage divider, using the load or ***I-V** curve. It allows us to "map" the complete current v. voltage response of the source, and thus determine the voltage (thevenin) and current (Norton) equivalents. 

## The 1.5 V alkaline battery

Use the potentiometer to change the load resistance ($R_L$) and measure the voltage and current *flowing through the load*.

```julia
#Voltage entries
iv = [3 6;4 5;5 4]
```

```julia
scatter(iv[:,2],iv[:,1],label=false,xlabel="current (A)",ylabel="voltage (V)", markersize=6,markershape=:diamond,markercolor=:purple )
```

Now that we have the ***I-V*** data points we can determine the Thevenin and Norton equivalents by determining the load function. That is, the linear function that fits the data set.

```julia
using Polynomials
fV=fit(iv[:,1],iv[:,2],1)
plot!(fV, extrema(iv[:,1])...)
```
